The Problem
Given a "black box" function f that maps n-bit strings to n-bit strings, the function is guaranteed to have a special property: there exists some secret string s (called the "period") such that for any two inputs x and y:
f(x) = f(y) if and only if x ⊕ y ∈ {0ⁿ, s}
In other words, the function is "two-to-one" - every output has exactly two inputs that map to it, and those two inputs differ by XOR-ing with the secret string s.
The goal: Find the hidden string s. This might seem like an artificial problem, and honestly, it kind of is! Simon's problem doesn't have major practical applications.But it was a major steeping stone which techniques are used by shor algorithm in later years.
